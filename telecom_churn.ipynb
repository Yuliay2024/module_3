```python
# заполняем пропущенные значения
def impute_NA_with_avg(data,strategy='mean',NA_col=[]):
    data_copy = data.copy(deep=True)
    for i in NA_col:
        if data_copy[i].isnull().sum()>0:
            if strategy=='mean':
                data_copy[i] = data_copy[i].fillna(data[i].mean())
            elif strategy=='median':
                data_copy[i] = data_copy[i].fillna(data[i].median())
            elif strategy=='mode':
                data_copy[i] = data_copy[i].fillna(data[i].mode()[0])
    return data_copy  
```

Анализ выбросов. Определим индексы выбросов для каждого столбца. В данном случае выбираем пороговое значение для детектирования выбросов как утроенное значение межквартильного размаха (IQR) относительно 75 % и 25 % квантилей. Необходимо уточнить, что для случаев нормального распределения достаточно и 1,5-кратного IQR, поскольку диапазон [Q(25 %) - 1.5IQR; Q(75 %) + 1.5IQR] охватывает более 99 % данных. Но многие переменные в рассматриваемом датасете не имеют нормального распределения. Напротив, их распределение сильно скошено в одну из сторон, поэтому чтобы не относить к выбросам чрезмерно большое количество данные, в качестве предела (threshold) принято значение 3.

```python
def outlier_detect_IQR(data,column,threshold=3):
       
    IQR = data[column].quantile(0.75) - data[column].quantile(0.25)
    Lower_fence = data[column].quantile(0.25) - (IQR * threshold)
    Upper_fence = data[column].quantile(0.75) + (IQR * threshold)
    #para = (Upper_fence, Lower_fence)
    tmp = pd.concat([data[column]>Upper_fence,data[column]<Lower_fence],axis=1)
    outliers = tmp.any(axis=1)
    outlier_indices = data[outliers].index

    return outlier_indices


```

```python
columns_with_outliers = []

for column in X_train.select_dtypes(include=['float64']).columns:
    if len(outlier_detect_IQR(X_train, column, threshold=3)) > 0:
        print(f'Количество выбросов в данных в столбце {column}: {len(outlier_detect_IQR(X_train, column, threshold=3))}')
        print(f'Индексы выборосов в данных в столбце {column}: {outlier_detect_IQR(X_train, column, threshold=3)}', '\n')
        columns_with_outliers.append(column)
```

```python
def impute_outlier_with_avg(data,column,outlier_index,strategy='median'):
    data_copy = data.copy(deep=True)
    if strategy=='mean':
        data_copy.loc[outlier_index,column] = data_copy[column].mean()
    elif strategy=='median':
        data_copy.loc[outlier_index,column] = data_copy[column].median()
    elif strategy=='mode':
        data_copy.loc[outlier_index,column] = data_copy[column].mode()[0]           
    return data_copy
```

Корреляционная фильтрация 
Находим признаки, которые сильно коррелируют друг с другом, так как такие признаки могут дублировать информацию и приводить к избыточности.

```python
def corr_feature_detect(data,threshold=0.8):
    
    corrmat = data.corr()
    corrmat = corrmat.abs().unstack() 
    corrmat = corrmat.sort_values(ascending=False)
    corrmat = corrmat[corrmat >= threshold]
    corrmat = corrmat[corrmat < 1] 
    corrmat = pd.DataFrame(corrmat).reset_index()
    corrmat.columns = ['feature1', 'feature2', 'corr']
   
    grouped_feature_ls = []
    correlated_groups = []
    
    for feature in corrmat.feature1.unique():
        if feature not in grouped_feature_ls:
    
            correlated_block = corrmat[corrmat.feature1 == feature]
            grouped_feature_ls = grouped_feature_ls + list(
                correlated_block.feature2.unique()) + [feature]
    
            correlated_groups.append(correlated_block)
    return correlated_groups
```

SelectKBest - Выбор лучших признаков

```python
from sklearn.feature_selection import SelectKBest, SelectPercentile
from sklearn.feature_selection import mutual_info_classif,chi2
```

```python
#Mutual Information - показывает сколько информации о целевой переменной можно извлечь из признаков. 
#Чем выше взаимная информация, тем больше информации признак несет о целевой переменной
def mutual_info(X,y,select_k=10):

    if select_k >= 1:
        sel_ = SelectKBest(mutual_info_classif, k=select_k).fit(X,y)
        col = X.columns[sel_.get_support()]
        
    elif 0 < select_k < 1:
        sel_ = SelectPercentile(mutual_info_classif, percentile=select_k*100).fit(X,y)
        col = X.columns[sel_.get_support()]   
        
    else:
        raise ValueError("select_k должно быть положительным значением")
    
    return col
```

Метрики

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import roc_auc_score, roc_curve

def quality_report(prediction, actual):
    print("Accuracy: {:.3f}\nPrecision: {:.3f}\nRecall: {:.3f}\nf1_score: {:.3f}".format(
        accuracy_score(actual, prediction),
        precision_score(actual, prediction),
        recall_score(actual, prediction),
        f1_score(actual, prediction)
    ))


def plot_roc_curve(prob_prediction, actual):
    ### реализовать функцию для отрисовки 
    fpr, tpr, thresholds = roc_curve(actual, prob_prediction)
    auc_score = roc_auc_score(actual, prob_prediction)
    
    plt.plot(fpr, tpr, label='ROC curve ')
    plt.plot([0, 1], [0, 1])
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC AUC: {:.3f}'.format(auc_score))
    plt.show()
```

```python
# Обучаем модель RandomForestClassifier с разными значениями n_estimators и max_depth 
results = []
for md in max_depths:
    for n_estimators in n_estimators_list:
        model_rfc = RandomForestClassifier(n_estimators=n_estimators, max_depth=md, random_state=42)
        model_rfc.fit(X_train, y_train)
        y_pred = model_rfc.predict(X_test)
        cur = accuracy_score(y_test, y_pred)
        prec = precision_score(y_test, y_pred)
        recc = recall_score(y_test, y_pred)
        f1 = f1_score(y_test, y_pred)
        results.append((n_estimators, md, cur, prec, recc, f1 ))
```

```python
result_df_rfc = pd.DataFrame(results, columns=['n_estimators', 'max_depth', 'ccuracy', 'precision', 'recall', 'f1_score'])
result_df_rfc
```
